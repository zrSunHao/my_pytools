### 图片融合

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1、读取操作
bg = cv2.imread('C:/Users/20806/Desktop/book/test/bg6.png')
img = cv2.imread('C:/Users/20806/Desktop/book/test/test.png')

h, w = img.shape[:2]
bg = cv2.resize(bg, (w, h), interpolation=cv2.INTER_AREA)

# 2、图像混合  bg1,bg3,bg5可以，bg3最优
ouput = cv2.addWeighted(img, 0.7, bg, 0.3, 0)
cv2.imwrite('C:/Users/20806/Desktop/python/ouput6.png',ouput)
# # 3、图像显示
# plt.imshow(ouput[:, :, ::-1])
# plt.show()
```

```python
import cv2
import os

# 1、读取操作
bg = cv2.imread('C:/Users/20806/Desktop/book/test/bg3_1.png')
root = 'C:/Users/20806/Desktop/python/transformer/'
files = os.listdir(root)

for ii,file in enumerate(files):
    img = cv2.imread(root+file)
    h, w = img.shape[:2]
    img_bg = cv2.resize(bg, (w, h), interpolation=cv2.INTER_AREA)
    ouput = cv2.addWeighted(img, 0.7, img_bg, 0.3, 0)
    cv2.imwrite('C:/Users/20806/Desktop/python/transformer_1/'+file,ouput)
```

### pdf转图片

```python
import fitz

doc = fitz.open("C:/Users/20806/Desktop/spp.pdf")
doc.page_count

for ii in range(doc.page_count):
    page=doc.load_page(ii)
    trans = fitz.Matrix(4, 4)
    pix = page.get_pixmap(matrix=trans, alpha=False)
    pix.save(f"C:/Users/20806/Desktop/python/spp/{ii+1}.png")
```

```python
import os
import fitz  

img_path = 'C:/Users/20806/Desktop/python/transformer_1'
files = os.listdir(img_path)
print(files)
files.sort(key=lambda x: int(x[:-4])) # 文件名 按数字排序
print(files)

doc = fitz.open()

# 循环path中的文件，可import os 然后用 for img in os.listdir(img_path)实现
# 这里为了让文件以1，2，3的形式进行拼接，就偷懒循环文件名中的数字。
for i, f_name in enumerate(files):
	img_file = img_path + '/' + f_name
	imgdoc = fitz.open(img_file)
	pdfbytes = imgdoc.convert_to_pdf()
	pdf_name = str(i) + '.pdf'
	imgpdf = fitz.open(pdf_name, pdfbytes)
	doc.insert_pdf(imgpdf)
doc.save('C:/Users/20806/Desktop/python/transformer_1.pdf')
doc.close()
```
